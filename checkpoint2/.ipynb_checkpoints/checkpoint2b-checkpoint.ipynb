<h2> Feature Selection </h2>

_Use this space to modify or create features_

```python
df = pd.read_csv('../data/states_edu.csv')
for i in range(len(df['AVG_MATH_4_SCORE'])):
    df = df[df['AVG_MATH_4_SCORE'].isnull() == bool(False)]
for i in range(len(df['AVG_MATH_8_SCORE'])):
    df = df[df['AVG_MATH_8_SCORE'].isnull() == bool(False)]
for i in range(len(df['AVG_READING_4_SCORE'])):
    df = df[df['AVG_READING_4_SCORE'].isnull() == bool(False)]
for i in range(len(df['AVG_READING_8_SCORE'])):
    df = df[df['AVG_READING_8_SCORE'].isnull() == bool(False)]
df = df.reset_index()

max = df.at[0,'AVG_MATH_4_SCORE']
min = df.at[0,'AVG_MATH_4_SCORE']
for j in range(len(df['AVG_MATH_4_SCORE'])):
    current_num = df.at[j,'AVG_MATH_4_SCORE']
    if(current_num > max):
        max = current_num
    if(current_num < min):
        min = current_num
for k in range(len(df['AVG_MATH_4_SCORE'])):
    df.at[k,'AVG_MATH_4_SCORE'] = ((df.at[k,'AVG_MATH_4_SCORE']-min)/(max-min))

    
max = df.at[0,'AVG_MATH_8_SCORE']
min = df.at[0,'AVG_MATH_8_SCORE']
for j in range(len(df['AVG_MATH_8_SCORE'])):
    current_num = df.at[j,'AVG_MATH_8_SCORE']
    if(current_num > max):
        max = current_num
    if(current_num < min):
        min = current_num
for k in range(len(df['AVG_MATH_8_SCORE'])):
    df.at[k,'AVG_MATH_8_SCORE'] = ((df.at[k,'AVG_MATH_8_SCORE']-min)/(max-min))
    
max = df.at[0,'AVG_READING_4_SCORE']
min = df.at[0,'AVG_READING_4_SCORE']
for j in range(len(df['AVG_READING_4_SCORE'])):
    current_num = df.at[j,'AVG_READING_4_SCORE']
    if(current_num > max):
        max = current_num
    if(current_num < min):
        min = current_num
for k in range(len(df['AVG_READING_4_SCORE'])):
    df.at[k,'AVG_READING_4_SCORE'] = ((df.at[k,'AVG_READING_4_SCORE']-min)/(max-min))
    
max = df.at[0,'AVG_READING_8_SCORE']
min = df.at[0,'AVG_READING_8_SCORE']
for j in range(len(df['AVG_READING_8_SCORE'])):
    current_num = df.at[j,'AVG_READING_8_SCORE']
    if(current_num > max):
        max = current_num
    if(current_num < min):
        min = current_num
for k in range(len(df['AVG_READING_8_SCORE'])):
    df.at[k,'AVG_READING_8_SCORE'] = ((df.at[k,'AVG_READING_8_SCORE']-min)/(max-min))
    
Total_Scores = df['AVG_READING_8_SCORE'] + df['AVG_READING_4_SCORE'] + df['AVG_READING_4_SCORE'] + df['AVG_READING_8_SCORE']
df['TOTAL_AVE_SCORES'] = Total_Scores
```

Final feature list: **With the limited time, I only scrubbed data rows that were missing any data for the AVG_MATH_4_SCORE, AVG_MATH_8_SCORE, AVG_READING_4_SCORE, or AVG_READING_8_SCORE. This data is then standardized to make it easier to compare aross the tests and gauge performance. The data for TOTAL_REVENUE will also be used in the final analysis as the model will attempt to predict this given the standardized data of the 4 averaged scores. An Aggregate score is inculded, however this is mostly for visualization purposes. **

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.dropna()
X = df[['AVG_READING_8_SCORE', 'AVG_READING_4_SCORE', 'AVG_MATH_4_SCORE', 'AVG_MATH_8_SCORE']]
y = df.loc[X.index]['TOTAL_REVENUE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=.3, random_state=0)
```

<h2> Prediction </h2>

Chosen ML task: **REGRESSION**

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
print(model.intercept_)
print(model.coef_)
```

```python
y_pred = model.predict(X_test)
```

```python
# FOR CLASSIFICATION ONLY:
```

```python
# FOR REGRESSION ONLY: (pick a single column to visualize results)

# Results from this graph _should not_ be used as a part of your results -- it is just here to help with intuition. 
# Instead, look at the error values and individual intercepts.


col_name = 'AVG_MATH_4_SCORE'
col_index = X_train.columns.get_loc(col_name)

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")
plt.scatter(X_test[col_name], model.predict(X_test), color = "blue")

new_x = np.linspace(X_train[col_name].min(),X_train[col_name].max(),200)
intercept = model.predict([X_train.sort_values(col_name).iloc[0]]) - X_train[col_name].min()*model.coef_[col_index]
plt.plot(new_x, intercept+new_x*model.coef_[col_index])

plt.legend(['controlled model','true training','predicted training','predicted testing'])
plt.xlabel(col_name)
plt.ylabel('TOTAL_REVENUE')
```
